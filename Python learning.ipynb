```python
#Recursion example

def factorial(n):
    if n==1 or n==0:
        return 1
    return n*factorial(n-1)    
    
print(factorial(5))   
```

```python
k = input("Enter a number to find Fibonacci series up to that number: ")

def fibonacci(m):
    if m == 0:
        return 0
    elif m == 1:
        return 1    
    else:
        return fibonacci(m - 1) + fibonacci(m - 2)

print(fibonacci(int(k)))
```
